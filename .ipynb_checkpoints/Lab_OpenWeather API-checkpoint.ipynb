```python
from dotenv import load_dotenv
import os 
```

```python
load_dotenv()
```

```python
my_key = os.getenv("API_KEY")
```

```python
#Then lets start to extract the data available, following API documentation instructions
```

```python
import requests
```

```python
url = "https://api.openweathermap.org/data/3.0/onecall"
```

```python
response = requests.get(url)
set(response.headers)
```

```python
url = url
parameters = {"lat": "38.717363", "lon": "-9.302266", "exclude": "minutely,alerts", "units" : "metric", "appid": my_key}
```

```python
response = requests.get(
    url=url, 
    params=parameters, 
    )
```

```python
response.status_code
```

```python
import pandas as pd
```

```python
data = response.json()
#to check the data from the API
print(data)
```

<h1 style="color: #00BFFF;">Data Manipulation & Insights </h1>

```python
#Lets focus on the hourly data
hourly_data = data["hourly"]
len(hourly_data)
```

```python
df = pd.DataFrame(hourly_data)
```

```python
df #this df retrieves the next 2 days hourly data
```

```python
#from df we can see that the weather column is not as insightful as it could be. Lets try to retrieve only the description value:
for data in hourly_data:
    weather_list = data["weather"] 
```

```python
weather_dict = weather_list[0]
weather_dict  
```

```python
main_weather_list = []
for data in hourly_data:
    main_weather_list.append(weather_dict["description"])
main_weather_list
```

```python
len(main_weather_list)
```

```python
#Now we want to concatenate df with df_weather. Lets first make a copy of df
```

```python
df2 = df.copy()
```

```python
df_weather = pd.DataFrame(main_weather_list)
```

```python
df_final = pd.concat([df2,df_weather], axis=1)
```

```python
df_final.head(10)
```

```python
#Lets drop weather column and then replace the name "0" to "weather" to our new column
df_final = df_final.drop(columns = ["weather"])
df_final.rename(columns={0 : "weather"}, inplace=True)
df_final.head(0)
```
